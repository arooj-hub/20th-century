# Loading Dataset

```python
file_path = "20th_century_events_cleaned.txt"

with open(file_path, "r", encoding="utf-8") as file:
    text_data = file.read()
    
print(text_data[:1000])  # Print first 1000 characters to check the content
```

# Tokenization

```python
tokens = word_tokenize(text_data.lower())  # Convert to lowercase
words = [word for word in tokens if word.isalnum()]  # Remove punctuation

word_freq = Counter(words)
common_words = word_freq.most_common(10)

# Convert to DataFrame for visualization
df_common_words = pd.DataFrame(common_words, columns=["Word", "Frequency"])

# Plot a bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x="Word", y="Frequency", data=df_common_words, palette="viridis")
plt.title("Top 10 Most Common Words (Before Removing Stopwords)")
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

# Stopwords

```python
stop_words = set(stopwords.words("english"))
filtered_words = [word for word in words if word not in stop_words]

word_freq_filtered = Counter(filtered_words)
common_filtered_words = word_freq_filtered.most_common(10)

# Convert to DataFrame
df_filtered_words = pd.DataFrame(common_filtered_words, columns=["Word", "Frequency"])

# Plot again
plt.figure(figsize=(10, 5))
sns.barplot(x="Word", y="Frequency", data=df_filtered_words, palette="plasma")
plt.title("Top 10 Most Common Words (After Removing Stopwords)")
plt.xlabel("Word")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

 After removing stopwords, important words like "war," "history," "soviet," and "world" become more prominent, highlighting key themes in the text. Common function words that do not add meaning, such as "the" and "is," are eliminated, making the data more informative. This improves topic modeling and text analysis by focusing on meaningful content rather than frequently used but non-informative words.

# POS tags with TextBlob


```python
blob = TextBlob(text_data)
pos_tags = blob.tags  # Get POS tags

# Count top 10 POS tags
pos_freq = Counter(tag for _, tag in pos_tags)
top_10_pos = pos_freq.most_common(10)

# Convert to DataFrame
df_pos = pd.DataFrame(top_10_pos, columns=["POS", "Frequency"])

# Plot the POS frequency
plt.figure(figsize=(10, 5))
sns.barplot(x="POS", y="Frequency", data=df_pos, palette="coolwarm")
plt.title("Top 10 POS Tags in the Article")
plt.xlabel("Part of Speech (POS)")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

# Bar Plots for Nouns, Verbs, and Adjectives

```python
nouns = [word for word, tag in pos_tags if tag.startswith("NN")]
verbs = [word for word, tag in pos_tags if tag.startswith("VB")]
adjectives = [word for word, tag in pos_tags if tag.startswith("JJ")]

# Count top 15
top_nouns = Counter(nouns).most_common(15)
top_verbs = Counter(verbs).most_common(15)
top_adjectives = Counter(adjectives).most_common(15)

# Convert to DataFrames
df_nouns = pd.DataFrame(top_nouns, columns=["Noun", "Frequency"])
df_verbs = pd.DataFrame(top_verbs, columns=["Verb", "Frequency"])
df_adjectives = pd.DataFrame(top_adjectives, columns=["Adjective", "Frequency"])

# Plot Nouns
plt.figure(figsize=(10, 5))
sns.barplot(x="Noun", y="Frequency", data=df_nouns, palette="Reds")
plt.title("Top 15 Nouns")
plt.xticks(rotation=45)
plt.show()

# Plot Verbs
plt.figure(figsize=(10, 5))
sns.barplot(x="Verb", y="Frequency", data=df_verbs, palette="Blues")
plt.title("Top 15 Verbs")
plt.xticks(rotation=45)
plt.show()

# Plot Adjectives
plt.figure(figsize=(10, 5))
sns.barplot(x="Adjective", y="Frequency", data=df_adjectives, palette="Greens")
plt.title("Top 15 Adjectives")
plt.xticks(rotation=45)
plt.show()
```

The frequent nouns like "war," "Hitler," "Germany," and "Soviet" indicate a strong focus on major 20th-century historical events, particularly World War-related topics. The verbs "retrieved," "was," "became," and "led" suggest discussions about historical transitions, leadership, and significant actions. Adjectives such as "German," "Japanese," "nuclear," and "Soviet" highlight geopolitical themes, emphasizing national identities and technological advancements in warfare.

#  Count Mentions of Countries

```python
# Define a simple country lookup list
countries = ["United States", "Germany", "Russia", "France", "China", "Japan", "India", "United Kingdom", "Italy", "Canada"]

country_counts = {country: len(re.findall(rf"\b{country}\b", text_data, re.IGNORECASE)) for country in countries}

# Convert to DataFrame
df_countries = pd.DataFrame(country_counts.items(), columns=["Country", "Mentions"]).sort_values(by="Mentions", ascending=False)

# Plot country mentions
plt.figure(figsize=(12, 6))
sns.barplot(x="Country", y="Mentions", data=df_countries, palette="magma")
plt.title("Frequency of Country Mentions in the Text")
plt.xticks(rotation=45)
plt.show()
```

Germany, Japan, and the United States are mentioned the most, likely due to their major roles in World War I, World War II, and the Cold War. France, Italy, and Russia were also key players in these global conflicts and political shifts. The presence of China and India suggests discussions on decolonization, revolutions, and rising global influence in the later half of the 20th century.
